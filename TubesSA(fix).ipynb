```python
items = df["No"]
```

```python
def powerset(items):
  res = [[]]
  for item in items:
    newset = [r+[item] for r in res]
    res.extend(newset)
  return res
```

```python
print(powerset(items))
```

```python
def bruteForceCalori(sort_profit,kalori):
  arr = powerset(items)
  max = 0
  max_budget = 0
  max_profit = 0
  max_karbo = 0
 
  for i in range(len(arr)):
    total = 0
    budget = 0
    total_profit = 0
    total_kalori = 0
    total_karbo = 0
    for j in range(len(arr[i])):
      total += sort_profit.Kalori[arr[i][j]-1]
      budget += sort_profit.Harga[arr[i][j]-1]
      total_profit += sort_profit.Profit[arr[i][j]-1]
      total_karbo += sort_profit.Karbohidrat[arr[i][j]-1]
    if max < total and total <= kalori and budget <= 100000:
        max_profit = total_profit
        max = total
        max_budget = budget
        max_karbo = total_karbo
        makanan = []
        for k in range(j+1):
          makanan.append(sort_profit.Makanan[arr[i][k]-1])
  print("Himpunan Solusi:",makanan)
  print("Total Biaya:",max_budget)
  print("Total Profit",max_profit)

bruteForceCalori(sort_profit,2000)
```
